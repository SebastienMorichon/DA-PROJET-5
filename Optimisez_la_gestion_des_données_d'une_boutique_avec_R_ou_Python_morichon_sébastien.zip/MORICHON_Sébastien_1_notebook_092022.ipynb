## 3. Fusion des tables

```python
# Fusion des tables 
df = pd.merge(erp, liaison, how="inner",on="product_id")
df = pd.merge(df, web_subset, how="inner",on="sku")
df.head()
```

```python
# Affichage des colonnes
df.columns
```

```python
# Ré-organisation des colonnes
df = df[['product_id','sku', 'post_title','post_date_gmt', 'onsale_web', 'stock_quantity', 'stock_status', 'total_sales', 'price']]
df
```

```python
# Création de la colonne Chiffre d'Affaire
df["Chiffre d'Affaire"] = df["total_sales"] * df["price"]
df.head()
```

### 3.1 - Vérification des valeurs abérantes

```python
# Calcul de la moyenne des prix
moyenne = df["price"].mean()
moyenne
```

```python
# Calcul de la médianne des prix
mediane = df["price"].median()
mediane
```

```python
# Affichage des valeurs abérantes des prix
df.boxplot(column="price", vert=False)
```

### 3.1.1 - Valeurs abérantes : Z-Score 

```python
df.head()
```

```python
# Création de la colonne Z-Score basé sur le prix
df["Z-Score : Price"] = stats.zscore(df["price"])
df
```

```python
# Tri des valeurs par ordre croissant
df.sort_values("Z-Score : Price").head()
```

```python
# Graphique de densité du Z-Score
df.plot(kind="density", y="Z-Score : Price", figsize=(10,6))
```

```python
# Calcul du % de l'aire du Z-Score
df["% Area : Price"] = stats.norm.cdf(df["Z-Score : Price"])
df
```

```python
# Tri de la table 
df.sort_values("% Area : Price")
```

```python
# Récupération des valeurs abérantes. Supérieur à 99,7% de l'air du Z-Score
outlier = df[df["% Area : Price"] > 0.997].sort_values("price", ascending=False)
outlier.info()
```

```python
outlier
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=outlier["post_title"], height=outlier["price"], label="Prix")
plt.title("Prix des bouteilles 'Outlier : Z-Score'", fontsize=15)
plt.axhline(moyenne, color="r", linestyle="--", label="Moyenne des prix")
plt.xticks(rotation=90)
plt.legend()
plt.xlabel("Nom du produit", fontsize=15)
plt.ylabel("Prix", fontsize=15)
plt.grid(axis="y")

# Affichage des valeurs des données sur le graphique
values = outlier["price"].tolist()
for i in range(19):
    plt.text(i-0.3, values[i]+5, int(values[i]), fontsize=9)

plt.show()
```

### 3.1.2 - Valeurs abérantes : Inter-quartiles

```python
df
```

```python
#On calcule Q1

q1=df["price"].quantile(q=0.25)
q1
```

```python
#On calcule Q3

q3=df["price"].quantile(q=0.75)
q3
```

```python
#On calcule l'écart interquartile (IQR)

IQR=q3-q1
IQR
```

```python
#On calcule la borne inférieure à l'aide du Q1 et de l'écart interquartile

borne_inf = q1-1.5*IQR
borne_inf
```

```python
#On calcule la borne supérieure à l'aide du Q3 et de l'écart interquartile

borne_sup = q3 +1.5*IQR
borne_sup
```

```python
#Affichage des valeurs abérantes
df_sup=df[df["price"]>borne_sup]
df_sup = df_sup.sort_values("price", ascending=False)
```

```python
# Affichage des données abérantes inférieur à borne_inf
df_inf=df[df["price"]<borne_inf]
df_inf
#Il n'y a pas de valeurs abérantes inférieur à la borne inf (comme vu plus haut sur le graphique à moustache)
```

```python
df_sup.info()
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=df_sup["post_title"], height=df_sup["price"], label="Prix")
plt.title("Prix des bouteilles 'Outlier : Inter-quartiles'", fontsize=15)
plt.axhline(moyenne, color="r", linestyle="--", label="Moyenne des prix")
plt.xticks(rotation=90)
plt.legend()
plt.xlabel("Nom du produit", fontsize=15)
plt.ylabel("Prix", fontsize=15)
plt.grid(axis="y")

# Affichage des valeurs des données sur le graphique
values = df_sup["price"].tolist()
for i in range(32):
    plt.text(i-0.3, values[i]+5, int(values[i]), fontsize=9)

plt.show()
```

## Aller plus loin : Calcul des valeurs abérantes pour les stocks

```python
df
```

```python
# Création de la colonne Stock de base
df["Stock de base"] = df["stock_quantity"] + df["total_sales"]
df.sort_values("Stock de base", ascending=False)
```

```python
# Moyenne des stock
moyenne_stock = df["Stock de base"].mean()
```

```python
# Création de la colonne % de ventes des stocks

df["% de vente des Stocks"] = round((df["total_sales"] / df["Stock de base"])*100,2)
df.sort_values("% de vente des Stocks", ascending=False)
```

```python
# Création de la colonne Z-Score basé sur le prix
df["Z-Score : Stock"] = stats.zscore(df["Stock de base"])
df
```

```python
# Tri des valeurs par ordre croissant
df.sort_values("Z-Score : Stock").head()
```

```python
# Graphique de densité du Z-Score : Stock
df.plot(kind="density", y="Z-Score : Stock", figsize=(10,6))
```

```python
# Calcul du % de l'aire du Z-Score
df["% Area : Stock"] = stats.norm.cdf(df["Z-Score : Stock"])
df
```

```python
# Tri de la table 
df.sort_values("% Area : Stock")
```

```python
# Récupération des valeurs abérantes. Supérieur à 99,7% de l'air du Z-Score : Stock
outlier_stock = df[df["% Area : Stock"] > 0.997].sort_values("Stock de base", ascending=False)
outlier_stock.info()
```

```python
plt.figure(figsize=(10,6))
plt.bar(x=outlier_stock["post_title"], height=outlier_stock["Stock de base"], label="Stock")
plt.bar(x=outlier_stock["post_title"], height=outlier_stock["total_sales"], label="Vente")
plt.title("Prix des bouteilles 'Outlier'", fontsize=15)
plt.axhline(moyenne_stock, color="r", linestyle="--", label="Moyenne des stock")
plt.xticks(rotation=90)
plt.legend()
plt.xlabel("Nom du produit", fontsize=15)
plt.ylabel("Stock", fontsize=15)
plt.grid(axis="y")

# Affichage des valeurs des données sur le graphique
values = outlier_stock["Stock de base"].tolist()
for i in range(20):
    plt.text(i-0.3, values[i]+10, int(values[i]), fontsize=9)

plt.show()
```

### Le graphique ci-dessous nous montre les valeurs abérantes basées sur les stock. De plus, nous pouvons constater que les ventes de ces produits semble très faible au vu du stock. Il serait judicieux de se pencher sur une ré-évaluation des quantités de stock en fonction des produits.
